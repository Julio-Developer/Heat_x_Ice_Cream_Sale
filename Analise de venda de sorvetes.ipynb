## Analisando a relação do calor com a venda de sorvetes

#### Esse propjeto, precisa mostrar a relação do aumento de temperatura com o aumento de venda de sorvetes

- Deve ficar identica a esse: 



#### Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Colocando lendo o dataset

```python
df = pd.read_csv('icecream_sales new.csv')
```

#### Começando a construção do gráfico

```python
# Organizando os dado em ordem crescente
order = df.sort_values('Temperature', ascending=False)

# Construindo o gráfico solicitado

# Definindo o titulo
plt.title('Ice Cream Truck')

# Construindo o grafico e colocando estilo
plt.plot(order['Temperature'], 
         order['Sales'],
         color = 'red',
         marker = '.',
         linestyle = '--',
         label = 'SU$')

# Definindo o nome dos eixos
plt.xlabel('Temperature (fahrenheit)')
plt.ylabel('Sales (US$)')
plt.legend()

#Salvando o arquivo em um png
plt.savefig('grafico sorvetes.png')

# Mostrando todos os graficos
plt.show()
```
